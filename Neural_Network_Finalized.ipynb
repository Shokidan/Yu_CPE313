```python
import numpy as np
import random

#This is so that the randint is constantly the same
random.seed(420)

class neuron():
  def __init__ (self,input,weight):
    self.input=np.array(input)
    self.weight=np.array(weight)

    self.sum=(sum(self.input*self.weight))

# Activation Function
  def relu(self):
    return max(0.0, self.sum)
    print(relu)

#Function for Feed forward, we use the random command to generate a random number between the range
  def feedforward(self):
    weight=random.randint(1,10)
    return self.relu(),weight
```

Though a possible way to check if the formula in excel is working this is unrecommended as in a neural network we are working with multiple nodes at once thus making this tedious and time consuming

```python
#Input and Weight Randomly Generated from excel
#Node1
input1=[5,6,7,8,9]
weight1=[1,3,5,7,9]

#Container for the results which will be the new input for the following node
input2=[]
#Randomly generated weight
weight2=[]

#The individual nodes, based from the screenshot above, there are five(5) nodes present
node=neuron(input1,weight1)
node1=neuron(input1,weight1)
node2=neuron(input1,weight1)
node3=neuron(input1,weight1)
node4=neuron(input1,weight1)

#Being that the node uses the same data, this is just purely to check the output of one of the
#nodes.
print("Output of the first node:",node.relu())

#Each App
input2.append(node.feedforward()[0])
weight2.append(node.feedforward()[1])

input2.append(node1.feedforward()[0])
weight2.append(node1.feedforward()[1])

input2.append(node2.feedforward()[0])
weight2.append(node2.feedforward()[1])

input2.append(node3.feedforward()[0])
weight2.append(node3.feedforward()[1])

input2.append(node4.feedforward()[0])
weight2.append(node4.feedforward()[1])
```

```python
print(input2)

print(weight2)
```

```python
output=neuron(input2,weight2)

print(output.relu())
```
